```python
def rename_filename(base_dir, base_name):
    if base_name == 'train':
        dest_folder = 'dataset/clean_dataset/train_dataset'
    elif base_name == 'val':
        dest_folder = 'dataset/clean_dataset/val_dataset'
    else:
        dest_folder = 'dataset/clean_dataset/test_dataset'

    for folder in os.listdir(base_dir):
        for idx, filename in enumerate(os.listdir(os.path.join(base_dir, folder))):
            new_filename = f"{base_name}_{folder}_{idx}.png"
            print(filename)
            print(new_filename)
            os.rename(os.path.join(base_dir,folder, filename), (os.path.join(dest_folder, folder, new_filename)))
```

```python
rename_filename(folder_train, 'train')
```
